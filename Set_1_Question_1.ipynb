```python
import cv2
import numpy as np
import os
from google.colab.patches import cv2_imshow

# 1. Read the image
image_path = 'input.png'  # Upload your image to the same Colab environment
image = cv2.imread(image_path)

if image is None:
    print("Image not found. Please upload an image named 'input.jpg'.")
else:
    # 2. Get image information
    height, width, channels = image.shape
    image_size = os.path.getsize(image_path)  # Size in bytes

    print("Image Information:")
    print(f" - Dimensions: {width}x{height}")
    print(f" - Channels: {channels}")
    print(f" - Original File Size: {image_size / 1024:.2f} KB")

    # 3. Save a compressed copy and find compression ratio
    compressed_path = 'compressed.jpg'
    cv2.imwrite(compressed_path, image, [cv2.IMWRITE_JPEG_QUALITY, 50])  # Compress at 50% quality
    compressed_size = os.path.getsize(compressed_path)

    compression_ratio = image_size / compressed_size
    print(f" - Compressed File Size: {compressed_size / 1024:.2f} KB")
    print(f" - Compression Ratio: {compression_ratio:.2f}")

    # 4. Display the negative of the image
    negative_image = 255 - image
    print("Displaying Original Image:")
    cv2_imshow(image)
    print("Displaying Negative Image:")
    cv2_imshow(negative_image)
```
